```python
# Даны две функции:
# f(x) = x^3 - 50x и g(x) = -x^4 + 88x^2 - 241
# Требуется:
# 1. Найти точки пересечения этих функций.
# 2. Построить графики функций в одной системе координат
# 3. Построить график общей функции
```

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
f, g = Symbol('f(x)'), Symbol('g(x)')
```

```python
funcF = x**3 - 50*x
```

```python
funcG = - x**4 + 88*x**2 - 241
```

```python
funcE = funcF - funcG
```

```python
plot(funcF, funcG)
```

```python
plot(funcE)
```

```python
solve(funcE)
```

```python
inters = solve(funcE)
```

```python
inters_x = []
```

```python
for i in inters:
    z = round(i.n(10), 2)
    inters_x.append(z)
```

```python
inters_x
```

```python
inters_y = []
```

```python
for i in inters_x:
    y1 = funcF.subs(x, i)
    inters_y.append(y1)
```

```python
inters_y
```

```python
for i, (x, y) in enumerate(zip(inters_x, inters_y), start = 1):
    print(f'точка пересечения {i} : ({x}, {y})')
```
